```python
import numpy as np
import pandas as pd
```

```python
x=np.array([[3,7],[4,6],[5,5],[7,5],[8,4],[5,1],[8,3],[9,3],[3,7],[3,5],[2,4],[5,9]])
y=np.array([[3,7],[4,6],[5,5],[7,5],[8,4],[5,1],[8,3],[9,3],[3,7],[3,5],[2,4],[5,9]])
```

```python
dm=np.zeros((len(x),len(y)))
```

```python
for i in range(len(x)):
  for j in range(len(y)):
    dm[i][j]=np.sqrt(np.sum((x[i]-y[j])**2))
rn=['P{}'.format(i+1)for i in range(len(x))]
cn=['P{}'.format(i+1)for i in range(len(x))]

df=pd.DataFrame(dm,index=rn,columns=cn)
```

```python
df
```

```python
t=1.9
nearestpoint={}

for point in df.index:
  nearest=[col for col in df.columns if col != point and df.loc[point,col]>0 and df.loc[point,col]<t]
  nearestpoint[point]=nearest

for point,nearest in nearestpoint.items():
  print(f"Nearest point to '{point}' are : ",nearest)
```
