```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y_actual = np.array([2.1, 2.9, 3.8, 5.2, 5.9])

# Fit a polynomial of degree 1 (linear fit)
coefficients = np.polyfit(x, y_actual, 1)

# Generate the predicted values using the fitted polynomial
y_predicted = np.polyval(coefficients, x)

# Calculate R-squared (coefficient of determination)
residuals = y_actual - y_predicted
ss_residual = np.sum(residuals ** 2)
ss_total = np.sum((y_actual - np.mean(y_actual)) ** 2)
r_squared = 1 - (ss_residual / ss_total)

# Convert R-squared to percentage
accuracy_percentage = r_squared * 100

# Scatter plot of the data
plt.scatter(x, y_actual, label='Actual values')

# Plot the fitted line
plt.plot(x, y_predicted, color='red', label='Fitted line')

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Print the R-squared as accuracy in percentage
print("Accuracy (R-squared): {:.2f}%".format(accuracy_percentage))

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score

# Sample data
x = np.array([1, 2, 3, 4, 5])
y_actual = np.array([2.1, 2.9, 3.8, 5.2, 5.9])

# Fit a polynomial of degree 1 (linear fit)
coefficients = np.polyfit(x, y_actual, 1)

# Generate the predicted values using the fitted polynomial
y_predicted = np.polyval(coefficients, x)

# Calculate R-squared using scikit-learn
r_squared = r2_score(y_actual, y_predicted)

# Convert R-squared to percentage
accuracy_percentage = r_squared * 100

# Scatter plot of the data
plt.scatter(x, y_actual, label='Actual values')

# Plot the fitted line
plt.plot(x, y_predicted, color='red', label='Fitted line')

# Add labels and a legend
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

# Print the R-squared as accuracy in percentage
print("Accuracy (R-squared): {:.2f}%".format(accuracy_percentage))

# Show the plot
plt.show()
```
